```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import collections
import time
import csv
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 5000)
pd.options.display.max_rows = 999
from IPython.display import display, HTML
```

```python
driver = webdriver.Chrome()
driver.get("https://sfbay.craigslist.org/search/sso")

all_posts = driver.find_elements_by_class_name("result-row")

header = []
price = []
links = []
content = []

craig_title = driver.find_elements_by_class_name("result-row")
hrefs = driver.find_elements_by_xpath("//li[@class='result-row']/a")
description = driver.find_elements_by_class_name("result-title.hdrlnk")

for n in craig_title:
    t_text = n.text
    content.append(t_text)
    
for n in hrefs:
    links.append(n.get_attribute('href'))
cdf = pd.DataFrame({'Content': content})

for row in cdf.Content:
    p = re.findall(r'[$]\S*', row)
    price.append(p[0])

for n in description:
    h_text = n.text
    header.append(h_text)
    
cdf = pd.DataFrame({'Header' : header, 'Price' : price, 'Link': links, 'Content': content })
cdf.to_csv("craigslist.csv", sep=',')

driver.close()
```
